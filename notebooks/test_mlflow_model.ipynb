```python
mlflow.set_tracking_uri("http://mlflow:5000")
model_name = "gnn_recommender"

# Получаем последнюю версию модели
client = MlflowClient()
latest_versions = client.get_latest_versions(model_name)

if not latest_versions:
    raise ValueError(f"Модель {model_name} не найдена в реестре")

latest_version = latest_versions[0]
print(f"🔄 Загружаем модель {model_name}, версия {latest_version.version}...")

# Формируем URI модели
model_uri = f"models:/{model_name}/{latest_version.version}"

# Загружаем модель как pyfunc
pyfunc_model = mlflow.pyfunc.load_model(model_uri)

# Инициализируем наш рекоммендер
model = pyfunc_model.unwrap_python_model().model

print(f"✅ Модель успешно загружена (версия {latest_version.version})")
```
