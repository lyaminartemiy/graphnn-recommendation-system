```python
import mlflow
import torch
from datetime import datetime


def upload_model_artifacts():
    # Настройка клиента MLflow
    mlflow.set_tracking_uri("http://0.0.0.0:5000")
    mlflow.set_experiment("RecommendationModels2")

    with mlflow.start_run(run_name=f"model_{datetime.now().strftime('%Y%m%d_%H%M%S')}"):
        # Пример сохранения модели PyTorch
        dummy_model = torch.nn.Linear(10, 2)
        mlflow.pytorch.log_model(dummy_model, "model")

        # Сохранение дополнительных артефактов
        with open("artifacts/metadata.json", "w") as f:
            f.write('{"model_version": "1.0.0"}')
        mlflow.log_artifact("artifacts/metadata.json")

        # Логирование параметров
        mlflow.log_params({"hidden_dim": 64, "num_layers": 2, "learning_rate": 0.001})


if __name__ == "__main__":
    upload_model_artifacts()
```
